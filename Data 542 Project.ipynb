```python
#Shiqi Zhang, Tyler Stevenson, Zefeng Pei
import numpy as np
import pandas as pd
pd.set_option('max_colwidth', 600)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style='darkgrid', font_scale=1.6)
import spacy
import random
from gensim import models, corpora, similarities
from wordcloud import WordCloud


#import discussion json file to dataframe
disc_shar = pd.read_json('20231012_235320_discussion_sharings.json')
disc_shar = pd.json_normalize(disc_shar['Sources'])

disc_shar.head()
```

```python
#import issue json file to dataframe
issue_shar = pd.read_json('20231012_235128_issue_sharings.json')
issue_shar = pd.json_normalize(issue_shar['Sources'])
issue_shar.head()
```

```python
# Get the common authors between both DataFrames
common_authors = set(issue_shar['Author']) & set(disc_shar['Author'])

# Get the common repos between both DataFrames
common_repo = set(issue_shar['RepoName']) & set(disc_shar['RepoName'])
```

```python
# Create a new dataframe that has all the authors and repos that have a comment and issue
issue = issue_shar[(issue_shar['Author'].isin(common_authors)) | (issue_shar['RepoName'].isin(common_repo))]
disc = disc_shar[(disc_shar['Author'].isin(common_authors)) | (disc_shar['RepoName'].isin(common_repo))]
data_total = pd.concat([issue, disc], axis=0, ignore_index=True)
data_total.head()
```

```python
# Create corpus from the 'Body' column of the 'data_total' DataFrame
corpus_total = data_total['Body'].tolist()

# Function to preprocess text (convert to lowercase)
def preprocess_text(text_total):
    return text_total.lower()

# Preprocess each document in the corpus
processed_corpus_total = [preprocess_text(text_total) for text_total in corpus_total]

# Load the spacy model for text processing
nlp = spacy.load("en_core_web_sm")

# Function to tokenize and lemmatize text, while removing punctuation, special characters, and stop words
def process_text(entry_total):
    # Apply spaCy pipeline to the entry (tokenization, lemmatization, etc.)
    doc = nlp(entry_total)
    
    # Return lemmatized tokens, excluding punctuation, spaces, non-alphanumeric characters, and stop words
    return [token.lemma_ for token in doc if not token.is_punct and not token.is_space and token.is_alpha and not token.is_stop]

# Process each document in the corpus using the function defined above
token_corpus_total = [process_text(entry_total) for entry_total in processed_corpus_total]

# Build a dictionary from the processed corpus (mapping token to unique ID)
dct_total = corpora.Dictionary(token_corpus_total)

# Convert the processed corpus into a Bag-of-Words (BoW) representation
bow_total = [dct_total.doc2bow(i) for i in token_corpus_total]

# Train an LDA (Latent Dirichlet Allocation) model on the Bag-of-Words representation
lda_model_total = models.LdaModel(corpus=bow_total, num_topics=10, passes=10, alpha='auto', eta='auto', id2word=dct_total, random_state=0)

# Get the topics for the corpus
topics_total = list(lda_model_total.get_document_topics(bow_total, minimum_probability=0))

# Print the proportion of each topic in the corpus
for i in range(10):
    print(f"Topic {i} proportion: {100 * sum([pair[i][1] for pair in topics_total]) / len(topics_total):.1f}%")

# Print the top words for each topic in the LDA model
lda_model_total.print_topics()
```

```python
# Plot word clouds for each topic
word_cloud = WordCloud(background_color='white', max_words=10)
fig, axes = plt.subplots(3, 3, figsize=(20,15))

for i, ax in enumerate(axes.flatten()):
    fig.add_subplot(ax)
    topic_words = dict(lda_model.show_topic(i))
    word_cloud.generate_from_frequencies(topic_words)
    plt.gca().imshow(word_cloud, interpolation='bilinear')
    plt.gca().set_title(f'Topic {i}')
    plt.gca().axis('off')
plt.show()
```
