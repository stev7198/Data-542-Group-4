```python
import pandas as pd
import json
from collections import Counter

json_file_path = "./DevGPT/snapshot_20231012/20231012_232232_hn_sharings.json"

with open(json_file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

conversations_data = []

for source in data.get("Sources", []):
    for chat in source.get("ChatgptSharing", []):
        for conversation in chat.get("Conversations", []):
            conversations_data.append({
                "Prompt": conversation.get("Prompt", ""),
                "Answer": conversation.get("Answer", ""),
                "TokensOfPrompts": chat.get("TokensOfPrompts", 0),
                "TokensOfAnswers": chat.get("TokensOfAnswers", 0),
                "Model": chat.get("Model", ""),
                "DateOfConversation": chat.get("DateOfConversation", ""),
                "Title": chat.get("Title", "")
            })

df_conversations = pd.DataFrame(conversations_data) 

categories = {
    "Bug Report": ["error", "bug", "issue", "exception", "traceback", "not working", "crash"],
    "Feature Request": ["feature", "add", "support", "implement", "new functionality"],
    "Theoretical Question": ["why", "how", "explain", "difference", "theory"],
    "Code Help": ["code", "function", "class", "method", "best practice", "optimize"],
    "General Inquiry": ["what", "when", "can", "possible", "is it"]
}

def categorize_prompt(prompt):
    prompt_lower = prompt.lower()
    for category, keywords in categories.items():
        if any(keyword in prompt_lower for keyword in keywords):
            return category
    return "Other"

df_conversations["Category"] = df_conversations["Prompt"].apply(categorize_prompt)

category_counts = Counter(df_conversations["Category"])

df_category_counts = pd.DataFrame(category_counts.items(), columns=["Category", "Count"])

df_category_counts




```
